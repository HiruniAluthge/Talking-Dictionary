```python
from tkinter import *
from PIL import Image, ImageTk
import json
from difflib import get_close_matches
from tkinter import messagebox
import pyttsx3

talk=pyttsx3.init()
voice=talk.getProperty('voices')
talk.setProperty('voice',voice[1].id)

def search():
    data = json.load(open('dictionary.json'))
    word = entry.get()
    word = word.lower()
    if word in data:
        meaning = data[word]
        textarea.delete(1.0,END)
        for item in meaning:
            textarea.insert(END,item)
            
    elif len(get_close_matches(word,data.keys()))>0:
        close_match=get_close_matches(word,data.keys())[0]
        res=messagebox.askyesno('Confirm',f'Did you mean {close_match} instead?')
        if res==True:
            entry.delete(0, END)
            entry.insert(END,close_match)
            meaning=data[close_match]
            textarea.delete(1.0,END)
            for item in meaning:
                textarea.insert(END,item)
        else:
            messagebox.showerror('Error', 'The word doesnot exist, Please double check it')
            entry.delete(0,END)
            textarea.delete(1.0,END)
        
    else:
        messagebox.showinfo('Information', 'The word doesnot exist.')
        entry.delete(0,END)
        textarea.delete(1.0,END)
        
def clear():
    entry.delete(0,END)
    textarea.delete(1.0,END)

def audio():
    talk.say(entry.get())
    talk.runAndWait()

def meaningaudio():
    talk.say(textarea.get(1.0,END))
    talk.runAndWait()
             
root = Tk()
root.geometry('800x626')
root.title('Talking Dictionary')
root.configure(bg='#242424')
root.resizable(0, 0)

title = Label(root, text='Talking Dictionary', font=('Allura', 25, 'bold'), fg='#ECC602', bg='#242424')
title.place(x=280, y=0)

Enterword = Label(root, text='Enter Word', font=('castellar', 20, 'bold'), fg='#ECC602', bg='#242424')
Enterword.place(x=300, y=60)

entry = Entry(root, font=('Allura', 20, 'bold'), bg='Whitesmoke', justify='center', relief=GROOVE)
entry.place(x=270, y=110)
entry.focus_set() 

image_search = Image.open('search.png')
image_search = image_search.resize((80, 80), Image.ANTIALIAS)
photo_search = ImageTk.PhotoImage(image_search)

searchbutton = Button(root, image=photo_search, bd=0, bg='#242424', activebackground='Whitesmoke', command=search)
searchbutton.image = photo_search  # Keep a reference to prevent garbage collection
searchbutton.place(x=300, y=160)

image_mic = Image.open('mic.png')
image_mic = image_mic.resize((80, 80), Image.ANTIALIAS)
photo_mic = ImageTk.PhotoImage(image_mic)

micbutton = Button(root, image=photo_mic, bd=0, bg='#242424', activebackground='Whitesmoke', command=audio)
micbutton.image = photo_mic  # Keep a reference to prevent garbage collection
micbutton.place(x=430, y=160)

meaning = Label(root, text='Meaning', font=('castellar', 20, 'bold'), fg='#ECC602', bg='#242424')
meaning.place(x=330, y=260)

textarea = Text(root, font=('Allura', 18, 'bold'), width=27, height=5, bd=4, relief= GROOVE,wrap ='word')
textarea.place(x=240, y=300)

image_meaningmic = Image.open('meaningmic.png')
image_meaningmic = image_meaningmic.resize((100, 100), Image.ANTIALIAS)
photo_meaningmic = ImageTk.PhotoImage(image_meaningmic)

meaningmicbutton = Button(root, image=photo_meaningmic, bd=0, bg='#242424', activebackground='Whitesmoke', command=meaningaudio)
meaningmicbutton.image = photo_meaningmic  # Keep a reference to prevent garbage collection
meaningmicbutton.place(x=320, y=480)

image_clear = Image.open('clear.png')
image_clear = image_clear.resize((80, 80), Image.ANTIALIAS)
photo_clear = ImageTk.PhotoImage(image_clear)

clearbutton = Button(root, image=photo_clear, bd=0, bg='#242424', activebackground='Whitesmoke', command=clear)
clearbutton.image = photo_clear  # Keep a reference to prevent garbage collection
clearbutton.place(x=420, y=480)

root.mainloop()
```
